```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.ticker import PercentFormatter
sns.set()
```

```python
df_used_cars = pd.read_csv("bar_chart_data.csv")
```

```python
df_used_cars
```

```python
plt.figure(figsize = (9, 6))
plt.barh(y = df_used_cars["Brand"], 
        width = df_used_cars["Cars Listings"], 
        color = "midnightblue")
plt.xticks( fontsize = 13)
plt.yticks( fontsize = 13)
plt.title("Cars Listings by Brand", fontsize = 16, fontweight = "bold")
plt.xlabel("Number of Listings", fontsize = 13 )
plt.tight_layout()
plt.savefig("Used Cars Bar.png")
plt.show()
```
